```python
# renaming columns for easier reference
df = df.rename(columns={"hate_speech": 'hate', "offensive_language": 'offensive', "neither": 'neutral', "class": 'target',"tweet":'speech'})
df.head()
```

```python
# create hate and non-hate categories by combining offensive and neutral categories
df.target = df.target.replace([2], 1) 
df.target = df.target.replace([0, 1], [1, 0]) 
df.target.value_counts()
```

```python
# Data preprossecing
#Preprocessing of the speech
# Removing usernames
import regex as re
def remove_users(df, col):
    df[col] = df[col].apply(lambda x: re.sub(r'(RT\s@[A-Za-z]+[A-Za-z0-9-_]+)', '', str(x))) 
    df[col] = df[col].apply(lambda x: re.sub(r'(@[A-Za-z0-9-_]+)', '', str(x)))

remove_users(df, 'speech')
df.head()
```

```python
# removing Character References
def remove_charef(df, col):
    df[col] = df[col].apply(lambda x: re.sub(r'&[\S]+?;', '', str(x)))

remove_charef(df, 'speech')
df.head()
```

```python
# removing hashtags
def remove_hashtags(df, col):
    df[col] = df[col].apply(lambda x: re.sub(r'#', ' ', str(x)))

remove_hashtags(df, 'speech')
df.head()
```

```python
# removing URLS
def remove_links(df, col): 
    df[col] = df[col].apply(lambda x: re.sub(r'http\S+', '', str(x)))  # remove http links 
    df[col] = df[col].apply(lambda x: re.sub(r'bit.ly/\S+', '', str(x)))  # remove bit.ly links    
    
remove_links(df, "speech")
df.head()
```

```python
# Removing Punctuation
def remove_punctuation(df, col):    
    df[col] = df[col].apply(lambda x: re.sub(r'[^\w\s]', r'', str(x)))

remove_punctuation(df, 'speech')
df.speech.head(20)
```

```python
# lower casing : making all the characters lower case
def lowercasing(df, col):
    df[col] = df[col].apply(lambda x: " ".join(x.lower() for x in x.split()))

lowercasing(df, 'speech')
df.speech.head(20)
```

```python
# remove numbers or words with digits  
def remove_numerics(df, col):
    df[col] = df[col].apply(lambda x: re.sub(r'\w*\d\w*', r'', str(x)))
    

remove_numerics(df, 'speech')
df.speech.head(20)
```

```python
# removing white-space
def remove_whitespaces(df, col):
    df[col] = df[col].apply(lambda x: re.sub(r'\s\s+', ' ', str(x))) # remove any double or more whitespaces to single whitespaces
    df[col] = df[col].apply(lambda x: re.sub(r'(\A\s+|\s+\Z)', '', str(x))) # remove any leading and trailing whitespace

remove_whitespaces(df, 'speech')
df.speech.head(20)
```

```python
#important libraries for preprocessing using NLTK
import nltk
from nltk import word_tokenize, FreqDist
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('omw-1.4')
from nltk.tokenize import TweetTokenizer
```

```python
# Tokenization and Lemmatization : Returns tokenized representation of words
import gensim
def lemmatize(token):
    """Returns lemmatization of a token"""
    return WordNetLemmatizer().lemmatize(token, pos='v')

def tokenize(tweet):
    """Returns tokenized representation of words in lemma form excluding stopwords"""
    result = []
    for token in gensim.utils.simple_preprocess(tweet):
        if token not in gensim.parsing.preprocessing.STOPWORDS \
                and len(token) > 2:  # drops words with less than 3 characters
            result.append(lemmatize(token))
    return result

def tokenize_and_lemmatize(df, col):
    df[col] = df[col].apply(lambda x: tokenize(x))
    df.speech = df.speech.apply(lambda x: str(x)[1:-1])

tokenize_and_lemmatize(df, 'speech')
df.speech.head()
```

```python
# separate dataframe into respective classes
hate = df[df.target == 1]
non_hate = df[df.target == 0]
```

```python
# perform 75-25 training-test split 
from sklearn.model_selection import train_test_split
X_hate_train,X_hate_test, y_hate_train, y_hate_test = train_test_split(hate.speech,hate.target, test_size=0.25, random_state=42)
X_non_hate_train, X_non_hate_test, y_non_hate_train, y_non_hate_test = train_test_split(non_hate.speech,non_hate.target, test_size=0.25, random_state=42)
```

```python
# concatenate hate and non-hate dataframe to reform entire training dataset
X_train = pd.concat((X_hate_train, X_non_hate_train), ignore_index=True)
y_train = pd.concat((y_hate_train, y_non_hate_train), ignore_index=True)
train = pd.concat([X_train, y_train], axis=1)
train.head()
```
